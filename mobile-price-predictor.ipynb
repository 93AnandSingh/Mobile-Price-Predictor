```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder,LabelEncoder,OrdinalEncoder
from sklearn.metrics import r2_score,mean_absolute_error
```

```python
from sklearn.linear_model import LinearRegression,Ridge,Lasso
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor,GradientBoostingRegressor,AdaBoostRegressor,ExtraTreesRegressor
```

```python
# List of categorical columns to encode
label_encoder = LabelEncoder()

categorical_columns = ['Company', 'Model', 'Colour', 'Processor_brand']

# Apply label encoding to each categorical column
for column in categorical_columns:
    df[column] = label_encoder.fit_transform(df[column])

df
```

```python
df.corr()['Prize']
```

```python
X = df.drop(columns=['Prize'])
y = np.sqrt(df['Prize'])
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.15,random_state=2)
```

```python
lr = LinearRegression()
lr
```

```python
lr.fit(X_train,y_train)
```

```python
y_pred = lr.predict(X_test)
y_pred
```

```python
final_df = pd.DataFrame({
    'Actual': y_test,
    'Predicted': y_pred
})

print(final_df)
```

```python
print('R2 score',r2_score(y_test,y_pred))
print('MAE', mean_absolute_error(y_test,y_pred))
```

```python
correlation = final_df['Actual'].corr(final_df['Predicted'])
print(f'Correlation Coefficient: {correlation}')
```

```python
residuals = final_df['Actual'] - final_df['Predicted']
sns.histplot(residuals, kde=True)
plt.xlabel('Residuals')
plt.title('Residuals Distribution')
plt.show()
```

```python
plt.scatter(final_df['Actual'], final_df['Predicted'])
plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')
plt.title('Actual vs Predicted')
plt.plot([final_df['Actual'].min(), final_df['Actual'].max()], [final_df['Actual'].min(), final_df['Actual'].max()], 'r--')
plt.show()
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error

# Assuming df is your DataFrame
mae = mean_absolute_error(final_df['Actual'], final_df['Predicted'])
mse = mean_squared_error(final_df['Actual'], final_df['Predicted'])
rmse = mean_squared_error(final_df['Actual'], final_df['Predicted'], squared=False)

print(f'MAE: {mae}')
print(f'MSE: {mse}')
print(f'RMSE: {rmse}')
```

```python
from sklearn.model_selection import cross_val_score

# Assuming model is your machine learning model and X, y are your features and target
scores = cross_val_score(lr, X, y, cv=5)
print(f'Cross-Validation Scores: {scores}')
print(f'Mean Accuracy: {scores.mean()}')
```
