## Prize column is showing object data type which we have to change into integer or float for our analysis

# Problem with Prize Column
## It contains a comma (like "7,299") values. The presence of the comma makes it difficult for the function to parse the string as a numeric value.

# To fix this, you need to remove the commas from the strings in the 'Prize' column before converting them to numeric values. You can do this using the `str.replace()` method.

```python
# Remove commas from the 'Prize' column and then convert to numeric
df['Prize'] = pd.to_numeric(df['Prize'].str.replace(',', ''), errors='coerce')  # Added str.replace to remove commas
print(df.head())
```

```python
final_df = pd.DataFrame({
    'Actual': y_test,
    'Predicted': y_pred
})

print(final_df)
```

```python
correlation = final_df['Actual'].corr(final_df['Predicted'])
print(f'Correlation Coefficient: {correlation}')
```

```python
from sklearn.metrics import r2_score,mean_absolute_error
```

```python
 print('R2 score',r2_score(y_test,y_pred))
```

```python
import matplotlib.pyplot as plt

plt.scatter(final_df['Actual'], final_df['Predicted'])
plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')
plt.title('Actual vs Predicted')
plt.plot([final_df['Actual'].min(), final_df['Actual'].max()], [final_df['Actual'].min(), final_df['Actual'].max()], 'r--')
plt.show()
```

```python
import pandas as pd
from sklearn.metrics import mean_absolute_error, mean_squared_error

# Assuming df is your DataFrame
mae = mean_absolute_error(final_df['Actual'], final_df['Predicted'])
mse = mean_squared_error(final_df['Actual'], final_df['Predicted'])
rmse = mean_squared_error(final_df['Actual'], final_df['Predicted'], squared=False)

print(f'MAE: {mae}')
print(f'MSE: {mse}')
print(f'RMSE: {rmse}')
```
